```python
x_data=data.drop(['Y_house_price_of_unit_area','No'],axis=1)
x_data
```

```python
y_data=data['Y_house_price_of_unit_area']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
train_score=lr.score(x_train,y_train)
```

```python
train_score
```

```python
test_score=lr.score(x_test,y_test)
test_score
```

```python
new_prediction=lr.predict(x_test)
new_prediction
```

```python
pd.DataFrame({'new_pre':new_prediction,'actual_data':y_test})
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_squared_error(new_prediction,y_test)
```

```python
mean_absolute_error(new_prediction,y_test)
```

```python
##model score
r2_score(new_prediction,y_test)
```
