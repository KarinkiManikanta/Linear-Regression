```python
from sklearn.linear_model import LinearRegression
```

```python
slope,intercept,r_value,p_value,std_error=stats.linregress(x,y_data)
```

```python
slope
```

```python
intercept
```

```python
plt.figure(figsize=(16,10))
plt.scatter(x,y_data)
plt.plot(x,intercept+slope*x,'r')
plt.title("linear regression line for  TV and Sales ")
plt.xlabel("TV")
plt.ylabel("Sales")
```

```python
x=np.array([[1,2,3,4,5,6,7]])
y=np.array([2,4,16,8,10,12,14])
```

```python
len(x),len(y)
```

```python
plt.scatter(x,y)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x.reshape(-1,1),y)
```

```python
lr.score(x.reshape(-1,1),y)
```

```python
lr.predict([[10]])
```

```python
%matplotlib inline
plt.scatter(x,y)
plt.plot(x,lr.intercept_+lr.coef_*x,'r')
```
