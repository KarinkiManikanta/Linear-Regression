```python
import numpy as np
import pandas as pd
import pandas_profiling 
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
data=pd.read_csv('/home/manikanta/Downloads/boston_house_prices.csv',skiprows=1)
```

```python
data
```

```python
data.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data.columns
```

```python
x_data=data[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
```

```python
y_data=data['MEDV']
```

```python
LR=LinearRegression()
```

```python
LR.fit(x_data,y_data)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
#liner Regression
```

```python
x=data[['CRIM']]
```

```python
#linerRegression For Crim and medv
LR.fit(x.values.reshape(-1,1),y_data)
```

```python
plt.figure(figsize=(16,10))
plt.title("linear regression line for  CRIM and MEDV ")
plt.xlabel("CRIM")
plt.ylabel("MEDV")
plt.scatter(x,y_data)
plt.plot(x,LR.intercept_+LR.coef_*x,'r')
```

```python
LR.score(x.values.reshape(-1,1),y_data)
```

```python
LR.predict([[0.02729]]) #34.7 (Error rate is:10.67822437)
```
